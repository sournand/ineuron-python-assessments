```python
#Assign1_Task1_Ques4
def computeSphereVolume(radius):
    pi = 3.1415926535897931
    return (4.0/3.0) * pi * (radius**3)

dia = 12
sphereVolume = computeSphereVolume((dia/2.0))
print('The volume of the sphere is =', sphereVolume)
```

```python
#Assign1_Task2_Ques2
def patternCreator(maxChar):
    drawStr = ""
    for i in list(range(maxChar)) :
        maxStar = i + 1
        countStar = 1
        while countStar <= maxStar  :
            drawStr = drawStr + '* '
            countStar = countStar + 1
        drawStr = drawStr + '\n'
    for i in list(range(maxChar - 1))[::-1] :
        maxStar = i + 1
        countStar = 1
        while countStar <= maxStar  :
            drawStr = drawStr + '* '
            countStar = countStar + 1
        drawStr = drawStr + '\n'
    return drawStr 
     
maxChar = 5
patternString = patternCreator(maxChar)
print(patternString)
```
